```python
data=pd.read_excel("/kaggle/input/sasan-data/SASAN.xlsx")
data.rename(columns = {'PPV (mm/s)':'PPV'}, inplace = True) 
data
```

```python
def normalize(dataset):
    dataNorm=((dataset-dataset.min())/(dataset.max()-dataset.min()))*1
    
    return dataNorm
d1=normalize(data)
d1
```

```python
X = d1[['No. of holes','Hole. dia.(mm)','Hole depth(m)','Burden(m)','Spacing(m)','Top stemming(m)','Avg. Explosive Per hole(kg)','Distance(m)']]
y=d1.PPV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
X_test
```

```python
y_test
```

**Random Forest approach**

```python
m = RandomForestRegressor(n_estimators=1000, n_jobs=-1)
m.fit(X_train,y_train)
y_pred=m.predict(X_test)
```

```python
rss=((y_test-y_pred)**2).sum()
mse=np.mean((y_test-y_pred)**2)
print("Final rmse value is =",np.sqrt(np.mean((y_test-y_pred)**2)))
```

**XG Boost model**

```python

xgb_model = xgb.XGBRegressor( objective="reg:linear",random_state=42)
xgb_model.fit(X_train, y_train)
xgb_model.score(X_test,y_test)
```

*RMSE Value*

```python
y_pre=xgb_model.predict(X_test)
rss=((y_test-y_pre)**2).sum()
mse=np.mean((y_test-y_pre)**2)
print("Final rmse value is =",np.sqrt(np.mean((y_test-y_pre)**2)))
```

***R squared value (R^2)***

```python
r2= r2_score(y_test,y_pre)
r2
```
